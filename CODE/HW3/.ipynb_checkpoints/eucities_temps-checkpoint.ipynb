```python
fig, axes = plt.subplots(2,2,figsize=(40,8))


color = {'R':'tab:red', 'O':'tab:orange', 'B':'tab:blue'}
regionDat = df.get(['city','EU','coastline','latitude','temperature'])

part1 = regionDat.loc[(df['coastline']=='no') & (df['EU'] == 'no')]

part2 = regionDat.loc[(df['coastline']=='yes') & (df['EU'] == 'no')]

part3 = regionDat.loc[(df['coastline']=='no') & (df['EU'] == 'yes')]

part4 = regionDat.loc[(df['coastline']=='yes') & (df['EU'] == 'yes')]



p1 = part1['latitude']
p2 = part1['city']
ra = []
for i in range(len(p2)):
    ra += [str(0+i)]
part1.loc[part1['temperature'] > 10, 'color'] = 'R'
part1.loc[((part1['temperature'] > 6) & (part1['temperature'] <= 10)), 'color'] = 'O'
part1.loc[part1['temperature'] < 6, 'color'] = 'B'
axes[0][0].scatter(ra,p1.values, c = part1['color'].map(color))
axes[0][0].set_title('No EU No Coastline')
axes[0][0].set_xlabel('City')
axes[0][0].set_ylabel('Latitude')


p3 = part2['latitude']
p4 = part2['city']
ra = []
for i in range(len(p4)):
    ra += [str(0+i)]
part2.loc[part2['temperature'] > 10, 'color'] = 'R'
part2.loc[((part2['temperature'] > 6) & (part2['temperature'] <= 10)), 'color'] = 'O'
part2.loc[part2['temperature'] < 6, 'color'] = 'B'
axes[0][1].scatter(ra,p3.values, c = part2['color'].map(color))
axes[0][1].set_title('No EU Yes Coastline')
axes[0][1].set_xlabel('City')
axes[0][1].set_ylabel('Latitude')


p5 = part3['latitude']
p6 = part3['city']
ra = []
for i in range(len(p6)):
    ra += [str(0+i)]
part3.loc[part3['temperature'] > 10, 'color'] = 'R'
part3.loc[((part3['temperature'] > 6) & (part3['temperature'] <= 10)), 'color'] = 'O'
part3.loc[part3['temperature'] < 6, 'color'] = 'B'
axes[1][0].scatter(ra,p5.values, c = part3['color'].map(color))
axes[1][0].set_title('Yes EU No Coastline')
axes[1][0].set_xlabel('City')
axes[1][0].set_ylabel('Latitude')


p7 = part4['latitude']
p8 = part4['city']
ra = []
for i in range(len(p8)):
    ra += [str(0+i)]
part4.loc[part4['temperature'] > 10, 'color'] = 'R'
part4.loc[((part4['temperature'] > 6) & (part4['temperature'] <= 10)), 'color'] = 'O'
part4.loc[part4['temperature'] < 6, 'color'] = 'B'
axes[1][1].scatter(ra,p7.values, c = part4['color'].map(color))
axes[1][1].set_title('Yes EU Yes Coastline')
axes[1][1].set_xlabel('City')
axes[1][1].set_ylabel('Latitude')

plt.show()
```
